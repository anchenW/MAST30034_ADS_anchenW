```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
# read the initial dataset
yellow_10 = pd.read_csv(r"../raw_data/yellow_tripdata_2018-10.csv")
yellow_11 = pd.read_csv(r"../raw_data/yellow_tripdata_2018-11.csv")
yellow_12 = pd.read_csv(r"../raw_data/yellow_tripdata_2018-12.csv")
```

```python
# test data 
yellow_19_10 = pd.read_csv(r"../raw_data/yellow_tripdata_2019-10.csv")
yellow_19_11 = pd.read_csv(r"../raw_data/yellow_tripdata_2019-11.csv")
yellow_19_12 = pd.read_csv(r"../raw_data/yellow_tripdata_2019-12.csv")
```

```python
# combine together
yellow_data = yellow_10.append(yellow_11).append(yellow_12)
```

```python
yellow_data_19 = yellow_19_10.append(yellow_19_11).append(yellow_19_12)
```

```python
# remove missing value
yellow = yellow_data.dropna(axis=0, how='any')
```

```python
yellow19 = yellow_data_19.dropna(axis=0, how='any')
```

```python
yellow.shape
```

```python
yellow19.shape
```

```python
# remove 0 passenger and more than 9 passengers
new_yellow = yellow[(yellow["passenger_count"] > 0)&(yellow["passenger_count"] < 10)]

# remove 0 trip_distance
new_yellow = new_yellow[new_yellow["trip_distance"] != 0]

# initial fare amount is $2.5
new_yellow = new_yellow[new_yellow["fare_amount"] >= 2.5]

# total payment >= 2.5
new_yellow = new_yellow[new_yellow["total_amount"] >= 2.5]

# remove invalid payment type since the drivers only seek for valid rewarding rides
new_yellow = new_yellow[(new_yellow["payment_type"] == 1)|(new_yellow["payment_type"] == 2)|(new_yellow["payment_type"] == 5)]

# restrict the pick up time in the valid period
new_yellow = new_yellow[(new_yellow["tpep_pickup_datetime"]<"2018-12-31 23:59:59")&(new_yellow["tpep_pickup_datetime"]>"2018-10-01 00:00:00")]
```

```python
# remove 0 passenger and more than 9 passengers
new_yellow19 = yellow19[(yellow19["passenger_count"] > 0)&(yellow19["passenger_count"] < 10)]

# remove 0 trip_distance
new_yellow19 = new_yellow19[new_yellow19["trip_distance"] != 0]

# initial fare amount is $2.5
new_yellow19 = new_yellow19[new_yellow19["fare_amount"] >= 2.5]

# total payment >= 2.5
new_yellow19 = new_yellow19[new_yellow19["total_amount"] >= 2.5]

# remove invalid payment type since the drivers only seek for valid rewarding rides
new_yellow19 = new_yellow19[(new_yellow19["payment_type"] == 1)|(new_yellow19["payment_type"] == 2)|(new_yellow19["payment_type"] == 5)]

# restrict the pick up time in the valid period
new_yellow19 = new_yellow19[(new_yellow19["tpep_pickup_datetime"]<"2019-12-31 23:59:59")&(new_yellow19["tpep_pickup_datetime"]>"2019-10-01 00:00:00")]
```

```python
new_yellow
```

```python
distance_boxplot = new_yellow.boxplot(column = "trip_distance", rot = 30)
fig = distance_boxplot.get_figure()

fig.suptitle("Box Plot of Distance")
fig.show()
```

```python
# remove outliers by using IQR
# calculating IQR
# https://towardsdatascience.com/practical-implementation-of-outlier-detection-in-python-90680453b3ce
Q1 = new_yellow.quantile(0.25)
Q3 = new_yellow.quantile(0.75)
IQR = Q3 - Q1
clean_data = new_yellow[~((new_yellow < (Q1 - 3*IQR)) | (new_yellow > (Q3 + 3*IQR))).any(axis = 1)]
```

```python
Q1 = new_yellow19.quantile(0.25)
Q3 = new_yellow19.quantile(0.75)
IQR = Q3 - Q1
clean_data19 = new_yellow19[~((new_yellow19 < (Q1 - 3*IQR)) | (new_yellow19 > (Q3 + 3*IQR))).any(axis = 1)]
```

```python
distance_boxplot = clean_data.boxplot(column = "trip_distance", rot = 30)
fig = distance_boxplot.get_figure()

fig.suptitle("Box Plot of Distance")
fig.show()
fig.savefig("distance after cleaning outliers")
```

```python
# feature engineering, to separate the "tpep_pickup_datetime" into two new columns of date and time respectively
clean_data["time"] = clean_data["tpep_pickup_datetime"].str[10:13]+":00"
clean_data["date"] = clean_data["tpep_pickup_datetime"].str[5:10]
```
